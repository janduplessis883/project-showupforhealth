```python
full_df = pd.read_csv('~/code/janduplessis883/data-showup/data/output-data/training_data_no_features.csv')
```

```python
full_df['Appointment status'].value_counts()
```

```python
full_df.shape
```

```python
def encode_appointment_status(df):
    df['App_status_encoded'] = [0 if app_status == 'Did Not Attend' else 1 
                                     for app_status in df['Appointment status']]
    df = df.drop(columns='Appointment status')
    return df
full_df = encode_appointment_status(full_df)
```
